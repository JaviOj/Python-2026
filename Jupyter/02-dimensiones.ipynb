```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generando un vector aleatorio 

```python
vector  = np.random.randint(1,11,12)
vector
```

## Obtención dimensión de un arreglo

```python
np.shape(vector)
```

## Tranformar vector en matriz de 4x3

```python
matriz = vector.reshape(4,3)
matriz
```

## Creación de un grafico

```python
plt.plot(vector, label='Vector')
plt.title("Gráfico del Vector") # Titulo
plt.xlabel("Indice (Posición)") # Titulo eje y
plt.ylabel("Valores") # Titulo eje x

plt.grid(True) # Agregar grilla al grafico
plt.legend() # Muestra etiqueta

plt.show() # Permite visualizar grafico
```
