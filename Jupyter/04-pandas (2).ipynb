### **Desafío en Clases: Analizando Tendencias Temporales con Gráficos de Línea**

Hasta ahora hemos utilizado gráficos de barras y de torta para comparar categorías fijas y proporciones. Sin embargo, en la ciencia de datos y la informática, una de las preguntas más comunes es: **¿Cómo cambian las variables a lo largo del tiempo?** Para responder a esto, el tipo de gráfico estándar es el **Gráfico de Línea (Line Plot)**.

### **Objetivo:**
Tu misión como ingeniero/a es descubrir cómo ha evolucionado la calidad (Rating promedio) de los videojuegos a lo largo de los años en nuestro dataset.

---

### **Instrucciones Paso a Paso:**

#### **Paso 1: Preparación y Limpieza de Fechas**
La columna `Release Date` actualmente es tratada por Pandas como texto (`object`). Además, contiene algunos registros con texto como `'releases on TBD'`. 
* Utiliza la función `pd.to_datetime()` con el argumento `errors='coerce'` para convertir esta columna al tipo de dato correcto (`datetime`). *Nota: `errors='coerce'` transformará automáticamente los textos inválidos en valores nulos (`NaT`), evitando que tu código falle.*

#### **Paso 2: Extracción del Año**
Una vez que la columna sea de tipo `datetime`, crea una nueva columna en el DataFrame llamada `Year` que contenga únicamente el año de lanzamiento de cada videojuego. 
* *Pista: Puedes usar el accesor de fechas de Pandas: `dataframe['Columna_Fecha'].dt.year`.*

#### **Paso 3: Agrupación y Cálculo del Promedio**
Necesitamos conocer el rating promedio *por cada año*. 
* Utiliza el método `.groupby('Year')` sobre tu DataFrame, selecciona la columna `Rating` y calcula su promedio con `.mean()`.
* Asegúrate de guardar este resultado en una nueva variable (por ejemplo, `tendencia_años`) y aplica `.dropna()` al final para eliminar cualquier año que haya quedado como nulo.

#### **Paso 4: Construcción del Gráfico**
Utilizando `matplotlib.pyplot`, genera un gráfico de línea que cumpla con los siguientes requisitos:
1. Define un tamaño de figura adecuado (ejemplo: `figsize=(10, 5)`).
2. Grafica los datos utilizando `plt.plot()`. Usa los años como el eje X (que corresponden al `.index` de tu agrupación) y los ratings promedio como el eje Y (los `.values`).
3. **Personalización:** Añade marcadores en los puntos (ejemplo: `marker='o'`), un estilo de línea continuo (`linestyle='-'`), un color a tu elección (ejemplo: `'teal'`, `'royalblue'`, etc.) y activa la cuadrícula (`plt.grid(True)`).
4. **Etiquetado:** Todo gráfico debe incluir Título, etiqueta en el eje X ("Año de Lanzamiento") y etiqueta en el eje Y ("Rating Promedio").

---

### **Preguntas de Reflexión**
1. ¿A partir de qué año se concentra la mayor cantidad de datos estables en el gráfico? ¿Qué crees que ocurrió con los años anteriores?
2. Si observas caídas o picos abruptos en la línea, ¿se deberá a una baja general en la calidad de los juegos de ese año o a que hay muy pocos juegos registrados en el dataset para ese período particular? ¿Cómo lo comprobarías programando?

```python
# Convertir la fecha a formato datetime (manejando errores por textos como 'TBD')
datos_sin_nulos['Release Date'] = datos_sin_nulos['Release Date'] = pd.to_datetime(
    datos_sin_nulos['Release Date'], 
    format='%b %d, %Y', 
    errors='coerce'
)

# Extraer el año
datos_sin_nulos['Year'] = datos_sin_nulos['Release Date'].dt.year

# Agrupar por año y calcular el promedio del Rating (eliminando datos sin año válido)
tendencia_años = datos_sin_nulos.groupby('Year')['Rating'].mean().dropna()

plt.figure(figsize=(10, 5))
plt.plot(tendencia_años.index, tendencia_años.values, marker='o', linestyle='-', color='teal', linewidth=2)

plt.title('Evolución del Rating Promedio de Videojuegos por Año')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Rating Promedio')
plt.grid(True)
plt.show()
```
