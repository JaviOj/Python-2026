```python
vector = np.random.randint(5, 21, 12)
```

```python
plt.plot(vector, label='Vector')
plt.title("Gráfico del Vector")
plt.xlabel("Indice (Posición)") # agrega el título del eje x
plt.ylabel("Valores") # agrega el título del eje y

plt.grid(True) # para agregar una grilla al gráfico
plt.legend() # mostrar una etiqueta, que se obtiene de la propiedad label del método plot()

plt.show() #permite visualizar gráfico
```
