```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Función 1
def procesar_caudal():
    # Generar horas desde 1 hasta 8 usando NumPy
    horas = np.arange(1, 9)
    
    # Lista con los caudales entregados
    caudales = [12.5, 14.2, 18.1, 22.8, 25.4, 21.0, 17.5, 15.3]
    
    # Crear DataFrame
    df = pd.DataFrame({
        'Hora': horas,
        'Caudal': caudales
    })
    
    return df
```

```python
# Función 2
def visualizar_curva(dataframe):
    fig, ax = plt.subplots()

    ax.plot(
        dataframe['Hora'],
        dataframe['Caudal'],
        marker='o',
        color='green',
        linestyle='--'
    )

    ax.set_title('Monitoreo de Caudal Río Gamboa')
    ax.set_xlabel('Tiempo (Horas)')
    ax.set_ylabel('Caudal (m3/s)')

    ax.grid(True)
    fig.savefig("grafico01",dpi=300)
```

```python
df_hidrologia = procesar_caudal()
visualizar_curva(df_hidrologia)
plt.show()
```
