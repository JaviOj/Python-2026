```python
sin_null= sin_duplicados.dropna()
sin_null
```

ejercicio letra d)

```python

def filtrar_fuego(sin_null):
    if 'Type 1' in sin_null.columns:
        return sin_null[sin_null['Type 1'] == 'Fire']
    tipo = sin_null.apply(lambda r: 'fire' in ' '.join(map(str, r)).lower(), axis=1)
    return sin_null[tipo]

fuego = filtrar_fuego(pokemon)
print("Total Pokémon tipo Fuego:", len(fuego))
print("Primeros 20 Pokémon de tipo Fuego:")
fuego.head(20)
```

ejercicio letra e)

```python
fuego["Def"].mean()
```

gradico de barra

```python

plt.bar(fuego["Name"].sort_values(ascending=False).head(10),fuego["Att"].sort_values(ascending=False).head(10),color=["cyan","olive","red","yellow","magenta","brown","orange","purple","blue","silver"])
plt.title("top 10 pokemon tipo fuego por nivel de ataque")
plt.xlabel("Nivel de defensa")
plt.ylabel("Pokemon")
plt.xticks(rotation=90)
plt.show()
```

```python
fuego.groupby(by="Generation")
```

```python
plt.plot(fuego["Generation"],marker="o")
plt.title("pokemon de tipo fuego por generacion")
plt.xlabel("cantidad")
plt.ylabel("generacion")
plt.grid(True)
plt.legend()
plt.show()
```
